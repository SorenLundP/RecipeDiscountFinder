```python
import requests
from bs4 import BeautifulSoup
import tkinter as tk
from tkinter import ttk

# Your recipes dictionary
recipes = {
    "Chilli_sin_carne": ["spidskommen", "kanel", "stødt korainder", "nudler"],
    "Pølsehorn": ["gær", "mælk", "smør", "rugbrød"],
    "Ost": ["bacon", "gifler", "ost", "brød"]
}

# Base URL for searching
base_url = "https://etilbudsavis.dk/soeg/{keyword}?business_ids=11deC%2CDWZE1w%2C9ba51"

# Dictionary to store offers for each ingredient
offers = {}

# Color codes for stores
store_colors = {
    'Rema 1000': 'blue',
    'Netto': 'yellow',
    '365discount': 'green',
}

# Function to fetch and store offers for an ingredient
def fetch_and_store_offers(ingredient):
    search_url = base_url.format(keyword=ingredient)
    response = requests.get(search_url)
    if response.status_code != 200:
        print(f"Failed to fetch offers for {ingredient}")
        return

    # Parse HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find offers
    offer_elements = soup.find_all('span', {'class': 'OfferList___StyledSpan2-sc-bj82vg-14'})

    # Extract and store the offers
    offers_for_ingredient = []
    for offer_element in offer_elements:
        offer_price = offer_element.text.replace('\xa0', ' ')
        offer_store_element = offer_element.find_parent().find('div', {'class': 'inline-flex'})
        offer_store = offer_store_element.text if offer_store_element else 'Unknown'

        offers_for_ingredient.append({
            'price': offer_price,
            'store': offer_store
        })

    if offers_for_ingredient:  # Only store if there are offers
        if ingredient in offers:
            offers[ingredient].extend(offers_for_ingredient)
        else:
            offers[ingredient] = offers_for_ingredient

# Loop through each recipe and its ingredients
for recipe, ingredients in recipes.items():
    print(f"Checking offers for {recipe}")

    # Count of discounted items for this recipe
    discount_count = 0

    for ingredient in ingredients:
        fetch_and_store_offers(ingredient)
        discount_count += len(offers.get(ingredient, []))

    print(f"Offers for {recipe} (only discounted items shown): {offers}")

    # Store the discount count for this recipe

# Rank recipes by the number of items on discount
recipe_ranking = {recipe: len(ingredients) for recipe, ingredients in recipes.items()}
sorted_recipes = sorted(recipe_ranking.items(), key=lambda x: x[1], reverse=True)

print("Recipes ranked by most items on discount:")
for recipe, count in sorted_recipes:
    print(f"{recipe}: {count} items on discount")

# Function to populate the GUI with offers
def populate_gui(root, offers_dict):
    row_num = 0
    col_num = 0
    for i, (recipe, ingredients) in enumerate(sorted_recipes):
        frame = ttk.Frame(root, padding="10")  # Removed the style
        frame.grid(row=row_num, column=col_num, padx=10, pady=10, sticky='nsew')

        label = ttk.Label(frame, text=f"{recipe} (Discounted Items)", font=("Arial", 16))  # Removed the style
        label.grid(row=0, column=0, columnspan=2, sticky='w')

        for ingredient in recipes[recipe]:
            if ingredient in offers_dict:
                ingredient_label = ttk.Label(frame, text=f"Item: {ingredient}", font=("Arial", 14))  # Removed the style
                ingredient_label.grid(row=row_num + 1, column=0, sticky='w')

                for discount in offers_dict[ingredient]:
                    store_label = ttk.Label(frame, text=f"Store: {discount['store']}", font=("Arial", 14), foreground=store_colors.get(discount['store'], 'black'))  # Removed the style
                    store_label.grid(row=row_num + 1, column=1, sticky='w')
                    row_num += 1

        if i % 2 == 0:  # Change the column for every 2 recipes
            col_num += 1
            row_num = 0

# Initialize tkinter root window
root = tk.Tk()
root.title("Discounted Recipe Ingredients")

# Populate GUI
populate_gui(root, offers)

root.mainloop()
```
